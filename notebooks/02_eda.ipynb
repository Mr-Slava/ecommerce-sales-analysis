```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

#   
df = pd.read_csv("C:/Python Project/ecommerce-sales-analysis/data/processed/data_clean.csv")

df.head()

# -10   - 
top_countries = df['Country'].value_counts().head(10)

sns.barplot(x=top_countries.values, y = top_countries.index, color='lightgreen')
plt.title("-10    ")
plt.xlabel(' ')
plt.ylabel('')
plt.show()

#    
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

#   
daily_sales = df.groupby(df['InvoiceDate'].dt.date)['TotalPrice'].sum()

daily_sales.plot(color='green')
plt.title('   ')
plt.xlabel('')
plt.ylabel(' ')
plt.tight_layout()
plt.show()
```
